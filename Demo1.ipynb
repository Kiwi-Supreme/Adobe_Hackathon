<center><h1>DEMO CODE</h1></center>

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

<b>Load Excel files</b>

```python
df_market = pd.read_excel("market_research.xlsx")
df_trending = pd.read_excel("trending_topics.xlsx")
```

<b>Merge based on Product</b>

```python
df_merged = pd.merge(df_trending, df_market, left_on="Related Product", right_on="Product")
```

<b>Group by Product to calculate statistics</b>

```python
df_grouped = df_merged.groupby("Product").agg({
"Popularity Score": "mean",
"Price (INR)": ["mean", "min"]
}).reset_index()
```

<b>Rename columns</b>

```python
df_grouped.columns = ["Product", "Popularity Score", "Avg Price", "Min Price"]
```

<b>Create Target Column</b>

```python
#Suggested Price = 90% of Min Price
df_grouped["Suggested Price"] = df_grouped["Min Price"] * 0.90
```

<b>Train Regression Model</b>

```python
X = df_grouped[["Popularity Score", "Avg Price", "Min Price"]]
y = df_grouped["Suggested Price"]

model = LinearRegression()
model.fit(X, y)
```

<b>Predict Prices</b>

```python
df_grouped["Predicted Price"] = model.predict(X).round(2)
```

<b>Save Output</b>

```python
df_grouped.to_excel("predicted_prices.xlsx", index=False)
```

<b>Print final predictions</b>

```python
print(df_grouped[["Product", "Popularity Score", "Avg Price", "Min Price", "Suggested Price", "Predicted Price"]])
```
